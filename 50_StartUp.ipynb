```python
statedf = pd.get_dummies(df['State'], drop_first=True)
statedf
```

```python
df = df.drop('State', axis=1)
df
```

```python
resultdf= pd.concat([df, statedf], axis=1)
resultdf
```

# SPLIT DATA

```python
x= resultdf.drop('Profit', axis=1)
y = resultdf['Profit']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state= 35)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
model= lm.fit(X_train, y_train)
```

```python
y_predict= model.predict(X_test)
y_predict
```

```python
df= pd.DataFrame({'y_test':y_test,
            'y_predict' :y_predict,
            'difference': abs(y_predict-y_test)})
df
```

```python
from sklearn.metrics import mean_absolute_error
MSE =mean_absolute_error(y_test, y_predict)
MSE
```

```python
RMSE = np.sqrt(MSE)
RMSE
```

```python
model.score(X_train, y_train)
```

```python
print('Intercept of the model:\n',lm.intercept_)
print("="*50)
print('Coefficient of the line:\n',lm.coef_)
```
